# No of Directories having Images

```python
path = "/content/drive/MyDrive/Data/DataSet"

directories = [d for d in os.listdir(path) if os.path.isdir(os.path.join(path, d)) and d.lower() != '.ipynb_checkpoints']

num_directories = len(directories)

print(f'Total number of directories: {num_directories}')
print('List of directories:')
print(directories)
```

# No of Images in each Directory

```python
for directory in directories:
    directory_path = os.path.join(path, directory)
    num_images = len([f for f in os.listdir(directory_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    print(f'Directory: {directory}, Number of Images: {num_images}')
```

```python
directory_names = []
image_counts = []

for directory in directories:
    directory_path = os.path.join(path, directory)
    num_images = len([f for f in os.listdir(directory_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

    directory_names.append(directory)
    image_counts.append(num_images)

plt.figure(figsize=(10, 6))
plt.bar(directory_names, image_counts, color='skyblue')
plt.xlabel('Directory')
plt.ylabel('Number of Images')
plt.title('Number of Images in Each Directory')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
import os
class_counts = {}

for directory in directories:
    directory_path = os.path.join(path, directory)
    num_images = len([f for f in os.listdir(directory_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    class_counts[directory] = num_images

print('Class Distribution:')
for class_name, count in class_counts.items():
    print(f'{class_name}: {count} images')
```
